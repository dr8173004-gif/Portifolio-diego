```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.available # estilos de graficos disponiveis
```

```python
plt.style.use('tableau-colorblind10') # como usar o grafico selecionado acima.
```

```python
plt.rcParams['figure.figsize'] = (20,7) #Passando as dimensões do grafico.
```

```python
d = pd.read_csv('titanic_train.csv')
d.head()
```

```python
d.shape
```

```python
plt.plot(d.Age, '*-r')
plt.title('Informações dos passageiros')
plt.xlabel('Passageiro')
plt.ylabel('Idade')
plt.show()
```

```python
plt.plot(d.Age, '*-r')
plt.title('Informações dos passageiros', size=18)
plt.xlabel('Passageiro', size=18)
plt.ylabel('Idade', size=18)
plt.show()
```

#Plotando Grafico direto no pandas

```python
d.Age.plot()
plt.xlabel('Passageiro', size=18)
plt.ylabel('Idade', size=18)
plt.show()
```

```python
d.plot()
plt.show()
```

#Scatter plot

*   Grafico de dispersão



```python
plt.scatter(d['PassengerId'], d['Age'], color= 'b', marker = '^')
plt.rcParams['figure.figsize'] = (15,7)
plt.show()
```
